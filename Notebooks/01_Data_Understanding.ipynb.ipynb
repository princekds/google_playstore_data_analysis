# Understanding final relationships in complete Google Play Store Dataset (2026)

## Data Model

This is essentially a **star-schema-like design** where `apps` acts as the central table.

```
                         apps (Primary Table)
                               |
                               | app_id (common key)
          ----------------------------------------------
          |                                            |
          |                                            |
app_country_stats                             discovery_signals
```

---

### `apps` (Master Table)

**Granularity:** One row per app.

Contains:
- App metadata
- Developer information
- Pricing
- Ratings
- Downloads
- Categories
- Android requirements
- Description
- Media assets

This is the fact table for app-level analytics.

---

### `app_country_stats`

**Granularity:** One row per app per country.

Contains:
- Country
- Installs
- Ratings
- Reviews
- Country-specific score
- Price
- Free/Paid

This enables regional analysis.

---

### `discovery_signals`

**Granularity:** One row per discovery event.

Contains:
- Search keyword
- Collection
- Chart rank
- Country
- Discovery source
- Signal category

This tells how users found the app.

---

- Written reviews = 445 thousand+

## Discovery Dataset

### Discovery Dataset Overview

This dataset contains information about how apps were discovered on the Google Play Store. It tracks the source, category, keywords, ranking, and signals that helped identify an app.

### App Information

| Column | Example Value | Meaning |
|---|---|---|
| id | 1 | Unique row identifier in the dataset. It is not the app identifier. |
| app_id | com.zhiliaoapp.musically | Unique package name/identifier of the app on Google Play Store. |
| source | chart_page | The method/source through which the app was discovered. Here, it was found from a Play Store chart page. |

### Discovery Location Information

| Column | Example Value | Meaning |
|---|---|---|
| country | us | Country where the discovery information was collected (US = United States). |
| category | SOCIAL | Google Play Store category of the app. |
| keyword | NaN | Search keyword used to find the app. Missing because this app was discovered through charts, not search. |

### Collection & Ranking Information

| Column | Example Value | Meaning |
|---|---|---|
| collection | topselling_free | Type of chart/list where the app was discovered. |
| chart_rank | 1.0 | Position of the app in the chart. Rank 1 means it was the top-ranked app in that category/list. |

For this app:
- Collection = Top Selling Free Apps
- Rank = #1

### Discovery Signal Information

| Column | Example Value | Meaning |
|---|---|---|
| signal_key | com.zhiliaoapp.musically\|chart_page\|us\|SOCIAL... | Combined unique identifier created from multiple discovery attributes. |
| discovered_at | 2026-05-05 | Date when this app was discovered/collected. |

### Example Interpretation

For this row:
TikTok (com.zhiliaoapp.musically) was discovered in the United States through a Google Play Store chart page. It appeared in the Social category under the "top selling free" collection and ranked number 1. The discovery record was collected on May 5, 2026.
