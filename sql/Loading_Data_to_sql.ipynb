```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
apps_df = pd.read_csv(
    r"C:\Users\cashk\OneDrive\Desktop\Projects\Google_Playstore_project\Playstore_Data\cleaned\apps_clean.csv"
)

country_df = pd.read_csv(
    r"C:\Users\cashk\OneDrive\Desktop\Projects\Google_Playstore_project\Playstore_Data\cleaned\country_clean.csv"
)

discovery_df = pd.read_csv(
    r"C:\Users\cashk\OneDrive\Desktop\Projects\Google_Playstore_project\Playstore_Data\cleaned\discovery_clean.csv"
)
```

```python
username = "postgres"
password = "root"
host = "localhost"
port = "5432"
database = "google_playstore"

engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)
```

```python
apps_df.to_sql(
    "apps",
    engine,
    if_exists="replace",
    index=False
)


country_df.to_sql(
    "country_stats",
    engine,
    if_exists="replace",
    index=False
)


discovery_df.to_sql(
    "discovery_signals",
    engine,
    if_exists="replace",
    index=False
)
```
